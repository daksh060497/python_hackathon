# Challenge 1: Smallest Palindrome


Create a function called `palindrome(num)` that given a number `'num'`, finds the next smallest palindrome and returns it

For eg:
***
**Input**

```python
palindrome(123)
```

**Output**
```python
131
```
***
**Input**

```python
palindrome(1331)
```

**Output**
```python
1441
```


````python
#Code starts here
def check_palindrome(num1):
  num1 = [int(x) for x in str(num1)] 
  if len(num1) % 2 == 0:
    tot = int(len(num1))
    var1 = int(tot/2)
    for i in range(0,var1):
      if num1[i] != num1[-1*(i+1)]:
        return False
    return True
  else:
    tot = int(len(num1))
    var1 = int((tot-1)/2)
    for i in range(0,var1):
      if num1[i] != num1[-1*(i+1)]:
        return False
    return True

def palindrome(num):
  num += 1
  while check_palindrome(num) == False:
    num += 1
  return num

palindrome(1331)
#Code ends here        
````

# Challenge 2: Anagram Scramble

Create a function called `'a_scramble(str_1, str_2)'` that given the strings`'str_1'`,`'str_2'` returns `True` if a portion of `'str_1'` characters can be scrambled to match `'str2'`, else return `False`

For eg:
***
**Input**

```python
a_scramble("Tom Marvolo Riddle","Voldemort")
```

**Output**
```python
True
```
***
**Input**

```python
a_scramble("ticket","chat")
```

**Output**
```python
False
```


````python
#Code starts here

def a_scramble(str_1, str_2):
  str_1 = list(str_1.lower())
  str_2 = list(str_2.lower())
  for i in str_2:
    if i not in str_1:
      return False
  return True
a_scramble("Tom Marvolo Riddle","Voldemort")

#Code ends here
````

# Challenge 3: Fibonacci Check

Create a function called `check_fib(num)` that checks if the given number `'num'` is part of the [fibonacci sequence](https://en.wikipedia.org/wiki/Fibonacci_number) and returns `True` if it is, else `False`

For eg:
***
**Input**

```python
check_fib(145)
```

**Output**
```python
False
```
***
**Input**

```python
check_fib(377)
```

**Output**
```python
True
```




````python
#Importing header files
def check_fib(num):
  x = 0
  y = 1
  
  febo = [0,1]

  while y <= (num+1):
      z = x + y
      x = y
      y = z
      febo.append(z)
  if num in febo: return True
  else: return False

check_fib(8)

#Code starts here


#Code ends here
````

# Challenge 4: String Compression

Write a function `compress(word)` that given a word `word` returns the string with it's letters and how many times they occur continously together.

For eg:

***
**Input**

```python
a_scramble("abbs")
```

**Output**
```python
"a1b2s1"
```
Explanation: 1 occurence of `a`, followed by 2 occurences of `b`, followed by 1 occurence of `s`
***
**Input**

```python
compress("xxcccdex")
```

**Output**
```python
"x2c3d1e1x1"
```

Explanation: 2 occurences of `x`, followed by 3 occurences of `c`, followed by 1 occurence of `d`, followed by 1 occurence of `e`, followed by 1 occurence of `x`

````python
#Code starts here
def compress(word):
  list1 = list(word.lower())
  list2 = list(dict.fromkeys(list1))
  list3 = []
  var2 = ""
  for i in list2:
    var1 = 0
    list3.append(i)
    for j in list1:
      if i == j: var1 += 1
    list3.append(str(var1))
  for x in list3:
    var2 += x
  return var2
compress("xxcccdex")
#Code ends here
````

# Challenge 5: K-Distinct

Write a function `k_distinct(string,k)` that given a string `'string'` and number `'k'`, it checks whether the `'string'` has `'k'` distinct characters

For eg:
***
**Input**

```python
k_distinct('Messoptamia',8)
```

**Output**
```python
True
```
***
**Input**

```python
k_distinct('banana',4)
```

**Output**
```python
False
```


````python
#Code starts here
def k_distinct(string,k):
  list1 = list(string.lower())
  list2 = list(dict.fromkeys(list1))
  if len(list2) == k:
    return True
  else:
    return False
k_distinct('banana',4)
#Code ends here
````
